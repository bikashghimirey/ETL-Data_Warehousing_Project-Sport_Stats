# Main Table

```python
import requests
from pprint import pprint

URL = 'https://www.thesportsdb.com/api/v1/json/3/search_all_teams.php?l=English%20Premier%20League'

response = requests.get(URL)
print(response)
```

```python
import pandas as pd
df_teams=pd.DataFrame(response.json()['teams'])
```

```python
pd.set_option('display.max_columns', 70)
df_teams.head()
```

```python
df_teams_selected = df_teams[['idTeam', 'strTeam', 'strTeamShort', 'intFormedYear' \
                              ,'idVenue', 'strStadium', 'strKeywords', 'strLocation', 'intStadiumCapacity', 'strWebsite','strBadge','strLogo']].copy()
df_teams_selected.head(30)
```

```python
from tqdm import tqdm

df= df_teams_selected.copy()
stadium_costs = []

for venue_id in tqdm(df['idVenue']):
    url = f"https://www.thesportsdb.com/api/v1/json/3/lookupvenue.php?id={venue_id}"
    response = requests.get(url)
    data = response.json()

    # Extract stadium cost if available
    if data['venues']:
        stadium_cost = data['venues'][0].get('strCost', None)
    else:
        stadium_cost = None

    stadium_costs.append(stadium_cost)

# Add the stadium cost to the DataFrame
df['stadiumCost'] = stadium_costs
```

```python
df.head(20)
```

```python
# Convert to numeric
df['intFormedYear'] = pd.to_numeric(df['intFormedYear'], errors='coerce')
df['intStadiumCapacity'] = pd.to_numeric(df['intStadiumCapacity'], errors='coerce')

# Replace 'None' and empty strings with NaN
df.replace(['None', '', ' '], pd.NA, inplace=True)
```

```python
import pandas as pd
import numpy as np
import re

def convert_cost_to_number(value):
    if pd.isna(value) or value == 'None':
        return np.nan

    # Lowercase for standardization
    value = value.lower()

    # Extract numeric part
    number_match = re.search(r'[\d.,]+', value)
    if not number_match:
        return np.nan

    number = number_match.group().replace(',', '')

    try:
        number = float(number)
    except:
        return np.nan

    # Handle million / billion
    if 'million' in value:
        return int(number * 1_000_000)
    elif 'billion' in value:
        return int(number * 1_000_000_000)
    else:
        return int(number)

# Apply to the stadiumCost column
df['stadiumCostCleaned'] = df['stadiumCost'].apply(convert_cost_to_number)

```

```python
location_split = df['strLocation'].str.split(',', expand=True)
df['city'] = location_split[0].str.strip()
df['region'] = location_split[1].str.strip() if location_split.shape[1] > 1 else None
df['country'] = location_split[2].str.strip() if location_split.shape[1] > 2 else 'England'

df['country'] = 'England'
```

```python
df['Keywords'] = df['strKeywords'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else None)
```

```python
avg_cost = df['stadiumCostCleaned'].mean(skipna=True)
df['stadiumCostCleaned'].fillna(avg_cost, inplace=True)
```

#Last 5 Match Table

```python
import requests
import pandas as pd
from tqdm import tqdm

# Prepare an empty list to hold match records
match_data = []

# Loop through each team and fetch last 5 matches
for team_id in tqdm(df['idTeam']):
    url = f"https://www.thesportsdb.com/api/v1/json/3/eventslast.php?id={team_id}"
    response = requests.get(url)
    data = response.json()

    if data['results']:
        for match in data['results'][:5]:
            match_data.append({
                'idTeam': team_id,
                'strEvent': match.get('strEvent'),
                'strLeague': match.get('strLeague'),
                'strHomeTeam': match.get('strHomeTeam'),
                'strAwayTeam': match.get('strAwayTeam'),
                'intHomeScore': match.get('intHomeScore'),
                'intAwayScore': match.get('intAwayScore'),
                'strVenue': match.get('strVenue'),
                'dateEvent': match.get('dateEvent'),
                'strTime': match.get('strTime'),
                'strStatus': match.get('strStatus')
            })

# Convert to a DataFrame
df_last_5_matches = pd.DataFrame(match_data)
```

```python
df_last_5_matches.shape
```

```python
pd.set_option('display.max_columns', 70)
df_last_5_matches.head(20)
```

```python
df_last_5_matches['result'] = df_last_5_matches.apply(
    lambda row: 'Win' if row['intHomeScore'] > row['intAwayScore'] else
                'Loss' if row['intHomeScore'] < row['intAwayScore'] else
                'Draw', axis=1)
```
